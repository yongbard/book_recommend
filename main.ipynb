```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.sparse import csr_matrix
from sklearn.neighbors import NearestNeighbors

# Load data
def load_data():
    books_df = pd.read_csv('./dataset/books.csv', sep=';')
    ratings_df = pd.read_csv('./dataset/ratings.csv', sep=';')
    users_df = pd.read_csv('./dataset/users.csv', sep=';')
    users_df['Age'] = pd.to_numeric(users_df['Age'], errors='coerce')
    return books_df, ratings_df, users_df

# Print data info
def print_data_info(books_df, ratings_df, users_df):
    print("도서 데이터프레임:")
    print(books_df.info())
    print("\n평점 데이터프레임:")
    print(ratings_df.info())
    print("\n사용자 데이터프레임:")
    print(users_df.info())
    print("\n사용자 나이 통계:")
    print(users_df['Age'].describe())

# Analyze missing age data
def analyze_age_missing(users_df):
    age_missing = users_df['Age'].isnull().sum()
    total_users = len(users_df)
    print(f"\n나이 정보가 없는 사용자: {age_missing}명 ({age_missing/total_users:.2%})")
    print(f"나이 정보가 있는 사용자: {total_users - age_missing}명 ({(total_users - age_missing)/total_users:.2%})")

# Plot age distribution
def plot_age_distribution(users_df):
    plt.figure(figsize=(10, 5))
    sns.histplot(users_df['Age'].dropna(), kde=True, bins=50)
    plt.title('사용자 나이 분포')
    plt.xlabel('나이')
    plt.ylabel('사용자 수')
    plt.show()

# Create age group
def age_group(age):
    if pd.isnull(age):
        return 'Unknown'
    elif age <= 18:
        return '0-18'
    elif age <= 30:
        return '19-30'
    elif age <= 50:
        return '31-50'
    else:
        return '51+'

# Create age groups
def create_age_groups(users_df):
    users_df['AgeGroup'] = users_df['Age'].apply(age_group)
    return users_df

# Plot age group distribution
def plot_age_group_distribution(users_df):
    age_order = ['0-18', '19-30', '31-50', '51+', 'Unknown']
    plt.figure(figsize=(10, 5))
    sns.countplot(x='AgeGroup', data=users_df, order=age_order)
    plt.title('나이 그룹 분포')
    plt.xlabel('나이 그룹')
    plt.ylabel('사용자 수')
    plt.show()

# Analyze ratings by age group
def analyze_age_group_ratings(merged_df):
    age_order = ['0-18', '19-30', '31-50', '51+', 'Unknown']
    age_group_ratings = merged_df.groupby('AgeGroup')['Rating'].mean().reindex(age_order)
    print("\n나이 그룹별 평균 평점:")
    print(age_group_ratings)

    plt.figure(figsize=(10, 5))
    sns.barplot(x=age_group_ratings.index, y=age_group_ratings.values, order=age_order)
    plt.title('나이 그룹별 평균 평점')
    plt.xlabel('나이 그룹')
    plt.ylabel('평균 평점')
    plt.show()

# Create user-item matrix
def create_user_item_matrix(ratings_df):
    user_ids = ratings_df['User-ID'].astype('category').cat.codes
    item_ids = ratings_df['ISBN'].astype('category').cat.codes
    ratings = ratings_df['Rating'].values
    shape = (user_ids.max() + 1, item_ids.max() + 1)
    return csr_matrix((ratings, (user_ids, item_ids)), shape=shape)

# Split ratings into train and test sets
def train_test_split_ratings(ratings_df, test_size=0.2):
    return train_test_split(ratings_df, test_size=test_size, random_state=42)

# Train and evaluate model
def train_and_evaluate_model(ratings_df, n_neighbors=5):
    train_df, test_df = train_test_split_ratings(ratings_df)
    
    # 전체 데이터셋에 대한 사용자 및 아이템 ID 매핑 생성
    user_ids = ratings_df['User-ID'].astype('category').cat.codes
    item_ids = ratings_df['ISBN'].astype('category').cat.codes
    user_id_map = dict(zip(ratings_df['User-ID'], user_ids))
    item_id_map = dict(zip(ratings_df['ISBN'], item_ids))
    
    # 훈련 데이터로 사용자-아이템 행렬 생성
    train_user_ids = [user_id_map[uid] for uid in train_df['User-ID']]
    train_item_ids = [item_id_map[iid] for iid in train_df['ISBN']]
    train_ratings = train_df['Rating'].values
    shape = (len(user_id_map), len(item_id_map))
    user_item_matrix = csr_matrix((train_ratings, (train_user_ids, train_item_ids)), shape=shape)
    
    model = NearestNeighbors(metric='cosine', algorithm='brute', n_neighbors=n_neighbors, n_jobs=-1)
    model.fit(user_item_matrix)
    
    y_true = []
    y_pred = []
    
    for idx, row in test_df.iterrows():
        user_id = row['User-ID']
        item_id = row['ISBN']
        if user_id in user_id_map and item_id in item_id_map:
            user_id_index = user_id_map[user_id]
            item_id_index = item_id_map[item_id]
            
            user_ratings = user_item_matrix[user_id_index].toarray().flatten()
            item_ratings = user_item_matrix[:, item_id_index].toarray().flatten()
            
            similar_users = model.kneighbors(user_ratings.reshape(1, -1), return_distance=False)[0]
            user_pred = np.mean(item_ratings[similar_users])
            
            y_true.append(row['Rating'])
            y_pred.append(user_pred)
    
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    print(f"\n모델 성능:")
    print(f"RMSE: {rmse:.4f}")
    
    return model, user_item_matrix, user_id_map, item_id_map

# Get top N recommendations
def get_top_n_recommendations(model, user_item_matrix, user_id, n=10):
    user_ratings = user_item_matrix[user_id].toarray().flatten()
    similar_users = model.kneighbors(user_ratings.reshape(1, -1), return_distance=False)[0]
    similar_user_ratings = user_item_matrix[similar_users]
    item_pred = np.mean(similar_user_ratings, axis=0)
    item_pred[user_ratings.nonzero()] = 0  # 이미 평가한 아이템 제외
    top_items = item_pred.argsort()[-n:][::-1]
    return top_items

# Main function
def main():
    # Load data
    books_df, ratings_df, users_df = load_data()
    print_data_info(books_df, ratings_df, users_df)

    # Analyze and handle age data
    analyze_age_missing(users_df)
    plot_age_distribution(users_df)
    users_df = create_age_groups(users_df)
    plot_age_group_distribution(users_df)

    # Merge data
    merged_df = pd.merge(ratings_df, users_df[['User-ID', 'AgeGroup']], on='User-ID')

    # Analyze ratings by age group
    analyze_age_group_ratings(merged_df)

    # Train and evaluate model
    model, user_item_matrix, user_id_map, item_id_map = train_and_evaluate_model(ratings_df)

    # 샘플 사용자에 대한 추천을 가져올 때
    sample_user_id = ratings_df['User-ID'].sample(1).iloc[0]
    user_id_index = user_id_map[sample_user_id]
    recommendations = get_top_n_recommendations(model, user_item_matrix, user_id_index)

    print(f"\n사용자 {sample_user_id}에 대한 상위 10개 추천 도서:")
    isbn_mapping = dict(zip(item_id_map.values(), item_id_map.keys()))
    for item_id in recommendations:
        isbn = isbn_mapping[item_id]
        book_title = books_df[books_df['ISBN'] == isbn]['Title'].values[0]
        print(f"ISBN: {isbn}, 제목: {book_title}")

    print("\n결론:")
    print("1. 나이 데이터의 결측치를 나이 그룹으로 처리했습니다.")
    print("2. 'Unknown' 그룹을 통해 모든 사용자를 데이터셋에 유지했습니다.")
    print("3. 나이 그룹별 사용자 분포를 확인했습니다.")
    print("4. 간단한 협업 필터링 모델을 구현했습니다.")
    print("5. 나이 그룹별 평균 평점 차이를 관찰했습니다.")
    print("\n다음 단계:")
    print("- 추천 모델 개선 (예: 행렬 분해 기법 적용)")
    print("- 나이 그룹 정보를 추천 시스템에 반영")
    print("- 나이 그룹별 추천 결과 분석")
    print("- 추천 시스템 성능 향상을 위한 다른 특성 탐색")

if __name__ == "__main__":
    main()
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from scipy.sparse import csr_matrix
import pickle
import os

def load_data(sample_size=10000):
    books_df = pd.read_csv('./dataset/books.csv', sep=';')
    ratings_df = pd.read_csv('./dataset/ratings.csv', sep=';')
    users_df = pd.read_csv('./dataset/users.csv', sep=';')
    users_df['Age'] = pd.to_numeric(users_df['Age'], errors='coerce')
    
    unique_users = ratings_df['User-ID'].unique()
    unique_items = ratings_df['ISBN'].unique()
    
    if len(ratings_df) > sample_size:
        sampled_users = np.random.choice(unique_users, size=min(len(unique_users), sample_size // 10), replace=False)
        sampled_items = np.random.choice(unique_items, size=min(len(unique_items), sample_size // 10), replace=False)
        ratings_df = ratings_df[ratings_df['User-ID'].isin(sampled_users) & ratings_df['ISBN'].isin(sampled_items)]
    
    return books_df, ratings_df, users_df

def create_user_item_matrix(ratings_df):
    user_ids = ratings_df['User-ID'].astype('category').cat.codes
    item_ids = ratings_df['ISBN'].astype('category').cat.codes
    ratings = ratings_df['Rating'].values
    shape = (user_ids.max() + 1, item_ids.max() + 1)
    return csr_matrix((ratings, (user_ids, item_ids)), shape=shape)

def save_model(model, filename):
    with open(filename, 'wb') as f:
        pickle.dump(model, f)

def load_model(filename):
    with open(filename, 'rb') as f:
        return pickle.load(f)

def train_and_evaluate_model(ratings_df, model_filename='book_recommendation_model.pkl'):
    train_df, test_df = train_test_split(ratings_df, test_size=0.2, random_state=42)
    
    if os.path.exists(model_filename):
        print("기존 모델을 로드합니다.")
        user_item_matrix = load_model(model_filename)
    else:
        print("새로운 모델을 학습합니다.")
        user_item_matrix = create_user_item_matrix(train_df)
        save_model(user_item_matrix, model_filename)
    
    y_true = []
    y_pred = []
    
    user_id_map = dict(zip(ratings_df['User-ID'], ratings_df['User-ID'].astype('category').cat.codes))
    item_id_map = dict(zip(ratings_df['ISBN'], ratings_df['ISBN'].astype('category').cat.codes))
    
    for _, row in test_df.iterrows():
        user_id = user_id_map.get(row['User-ID'])
        item_id = item_id_map.get(row['ISBN'])
        
        if user_id is not None and item_id is not None and user_id < user_item_matrix.shape[0] and item_id < user_item_matrix.shape[1]:
            user_ratings = user_item_matrix[user_id].toarray().flatten()
            item_ratings = user_item_matrix[:, item_id].toarray().flatten()
            
            if np.sum(user_ratings != 0) > 0 and np.sum(item_ratings != 0) > 0:
                user_mean = np.mean(user_ratings[user_ratings != 0])
                item_mean = np.mean(item_ratings[item_ratings != 0])
                pred = (user_mean + item_mean) / 2
                
                y_true.append(row['Rating'])
                y_pred.append(pred)
    
    if len(y_true) > 0:
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        print(f"\n모델 성능:")
        print(f"RMSE: {rmse:.4f}")
    else:
        print("\n경고: 테스트 데이터에서 예측을 수행할 수 없었습니다.")
    
    return user_item_matrix, user_id_map, item_id_map

def get_top_n_recommendations(user_item_matrix, user_id, n=10):
    if user_id >= user_item_matrix.shape[0]:
        print(f"경고: 사용자 ID {user_id}가 행렬의 범위를 벗어났습니다.")
        return []
    
    user_ratings = user_item_matrix[user_id].toarray().flatten()
    unrated_items = np.where(user_ratings == 0)[0]
    item_means = np.mean(user_item_matrix[:, unrated_items].toarray(), axis=0)
    top_items = unrated_items[item_means.argsort()[-n:][::-1]]
    return top_items

def main():
    books_df, ratings_df, users_df = load_data(sample_size=5000000)
    print(f"샘플링된 평점 데이터 크기: {len(ratings_df)}")

    user_item_matrix, user_id_map, item_id_map = train_and_evaluate_model(ratings_df)

    sample_user_id = ratings_df['User-ID'].sample(1).iloc[0]
    user_id_index = user_id_map.get(sample_user_id)
    
    if user_id_index is not None:
        recommendations = get_top_n_recommendations(user_item_matrix, user_id_index)
        
        print(f"\n사용자 {sample_user_id}에 대한 상위 10개 추천 도서:")
        isbn_mapping = {v: k for k, v in item_id_map.items()}
        for item_id in recommendations:
            isbn = isbn_mapping.get(item_id)
            if isbn is not None and isbn in books_df['ISBN'].values:
                book_title = books_df[books_df['ISBN'] == isbn]['Title'].values[0]
                print(f"ISBN: {isbn}, 제목: {book_title}")
    else:
        print(f"경고: 사용자 ID {sample_user_id}를 찾을 수 없습니다.")

if __name__ == "__main__":
    main()
```
